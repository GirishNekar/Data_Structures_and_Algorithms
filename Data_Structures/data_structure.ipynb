<h1> Stack Data Structure <h1>

Implementation Of Stack

```python
class Stack:

    def __init__(self):

        self.items = []

    def push(self, item):
        
        """
        Description : Pushes an item onto the stack.

        Parameters:
            item (any): The item to be pushed onto the stack.

        Returns:
            None
        """
        
        self.items.append(item)




    def pop(self):
        
        
        """
        Description : Removes and returns the top item from the stack.

        Parameter : None
        
        Returns:
            any: The top item from the stack if the stack is not empty.
            None: If the stack is empty.
        """
        
        if not self.is_empty():
            return self.items.pop()
        return None



    def peek(self):
        
        
        """
        Description : Returns the top item from the stack without removing it.

        Parameter : none
        
        Returns:
            any: The top item from the stack if the stack is not empty.
            None: If the stack is empty.
        """
        
        
        if not self.is_empty():
            return self.items[-1]
        return None
```

1.Stack Operations



```python

class Stack:

    def __init__(self):

        self.items = []

    def push(self, item):
        
        
        """
        Description : Pushes an item onto the stack.

        Parameters:
            item (any): The item to be pushed onto the stack.

        Returns:
            None
        """
        
        
        self.items.append(item)



    def pop(self):
        
        
        """
        Description : Removes and returns the top item from the stack.

        Parameter : None
        
        Returns:
            any: The top item from the stack if the stack is not empty.
            None: If the stack is empty.
        """
        
        if not self.is_empty():
            return self.items.pop()
        return None


    def peek(self):
        
        
        """
        Description : Returns the top item from the stack without removing it.

        Parameter : none
        
        Returns:
            any: The top item from the stack if the stack is not empty.
            None: If the stack is empty.
        """
        
        
        if not self.is_empty():
            return self.items[-1]
        return None



    def is_empty(self):
        
        
        """
        Description : Checks if the stack is empty.
        
        Parameter : None

        Returns:
            bool: True if the stack is empty, False otherwise.
        """
        
        return len(self.items) == 0



    def size(self):
        
        
        """
        Description : Returns the size of the stack.

        Parameter : None
        
        Returns:
            int: The number of items in the stack.
        """
        
        
        return len(self.items)



if __name__ == "__main__":
    stack = Stack()
    stack.push(1)
    stack.push(2)
    stack.push(3)
    stack.push(4)
    
    print("Popped item:", stack.pop()) 
    print("Top item after pop:", stack.peek())  
    print("Is stack empty?", stack.is_empty())  
```

2.Paranthesis Check

```python

def is_balanced(s):
    
    """
    Check if the parentheses in the string are balanced.
    
    Parameters:
        s (str): The string containing parentheses.
    
    Returns:
        bool: True if balanced, False otherwise.
    """
    
    stack = Stack()
    pairs = {')': '(', '}': '{', ']': '['}
    
    
    for char in s:
        if char in '({[':
            stack.push(char)
        elif char in ')}]':
            if stack.is_empty() or stack.pop() != pairs[char]:
                return False
    
    return stack.is_empty()


def main():
    
    print(is_balanced("({[]})"))  
    print(is_balanced("({[})"))   


if __name__ == "__main__":
    main()
```

3. Sorting The Stack

```python

def sort_stack(stack):
    
    """
    Description :
    This function sorts a stack.

    Parameters:
        stack (Stack): The stack to be sorted.

    Returns:
        None
    """
    
    temp_stack = Stack()

    while not stack.is_empty():
        current = stack.pop()

        while not temp_stack.is_empty() and temp_stack.peek() > current:
            stack.push(temp_stack.pop())

        temp_stack.push(current)

    while not temp_stack.is_empty():
        stack.push(temp_stack.pop())


def main():
    stack = Stack()
    stack.push(3)
    stack.push(1)
    stack.push(4)
    stack.push(2)
    sort_stack(stack)
    
    
    print("Sorted stack:")
    
    while not stack.is_empty():
        print(stack.pop())  
         
if __name__ == "__main__":
    main()
```

<h1> Queue Data Structure <h1>

1. Queue Implementation with Basic Operations

```python


class Queue:
    
    """
    A class to represent a queue data structure.

    Attributes:
        items (list): A list to store queue elements.

    Methods:
        enqueue(item): Adds an item to the rear of the queue.
        dequeue(): Removes and returns the item from the front of the queue.
        peek(): Returns the front item of the queue without removing it.
        is_empty(): Checks if the queue is empty.
        size(): Returns the size of the queue.
    """

    def __init__(self):
       
        self.items = []



    def enqueue(self, item):
        
        
        """
        Description : 
            Adds an item to the rear of the queue.

        Parameters:
            item (any): The item to be added to the queue.

        Returns:
            None
        """
        
        
        self.items.append(item)


    def dequeue(self):
        
        
        """
        Description : 
            Removes and returns the item from the front of the queue.

        Parameters :
            None
        
        Returns:
            any: The front item from the queue if the queue is not empty.
            None: If the queue is empty.
        """
        
        
        if not self.is_empty():
            return self.items.pop(0)
        return None


    def peek(self):
        
        
        """
        Description :Returns the front item of the queue without removing it.
        
        Parameters :
            None
            
        Returns:
            any: The front item of the queue if the queue is not empty.
            None: If the queue is empty.
        """
        
        
        if not self.is_empty():
            return self.items[0]
        return None


    def is_empty(self):
        
        
        """
        Description :Checks if the queue is empty.

        Parameters :
            None
            
        Returns:
            bool: True if the queue is empty, False otherwise.
        """
        
        
        return len(self.items) == 0


    def size(self):
        
        
        """
        Description :Returns the size of the queue.

        Parameters :
            None
            
        Returns:
            int: The number of items in the queue.
        """
        
        
        return len(self.items)



def main():
    queue = Queue()
    queue.enqueue(1)
    queue.enqueue(2)
    queue.enqueue(3)

    print("Front item:", queue.peek()) 
    print("Dequeued item:", queue.dequeue()) 
    print("Is queue empty?", queue.is_empty())  
    print("Queue size:", queue.size())


if __name__ == "__main__":
    
    main()
  
```

2. Reverse A Queue

```python


def reverse_queue(queue):
    
    """
    Description : 
        Reverses the elements of a queue.

    Parameters:
        queue (Queue): The queue to be reversed.

    Returns:
        None
    """
    
    if not queue.is_empty():
        item = queue.dequeue()
        reverse_queue(queue)
        queue.enqueue(item)



def main():
    
    queue = Queue()
    queue.enqueue(1)
    queue.enqueue(2)
    queue.enqueue(3)
    queue.enqueue(4)

    print("Original Queue:")
    while not queue.is_empty():
        print(queue.dequeue(), end=" ")  
    print()


    queue.enqueue(1)
    queue.enqueue(2)
    queue.enqueue(3)
    queue.enqueue(4)

   
    reverse_queue(queue)

    print("Reversed Queue:")
    while not queue.is_empty():
        print(queue.dequeue(), end=" ")  
    print()


if __name__ == "__main__":
    main()

```

3. Generate Binary Numbers Using a Queue

```python


def generate_binary_numbers(n):
    
    
    """
    Dscription : 
        Generates the first n binary numbers using a queue.

    Parameters:
        n (int): The number of binary numbers to generate.

    Returns:
        list: A list containing the first n binary numbers as strings.
    """
    
    
    
    result = []
    queue = Queue()

    queue.enqueue("1")

    for _ in range(n):
        front = queue.dequeue()
        result.append(front)

        queue.enqueue(front + "0")
        queue.enqueue(front + "1")

    return result



def main():
    n = 15
    binary_numbers = generate_binary_numbers(n)
    print("First", n, "binary numbers:", binary_numbers)

if __name__ == "__main__":
    main()

```


<h1> Trees Data Structure <h1>

1.Tree Implementation 

```python

class Node:
    
    
    """
    Description:
        A class to represent a node in a binary tree.

    Attributes:
        data (any): The data stored in the node.
        left (Node): The left child of the node.
        right (Node): The right child of the node.
    """

    def __init__(self, data):
        
        
        """
        Description:
            Initializes a binary tree node with the given data and no children.

        Parameters:
            data (any): The data to be stored in the node.

        Returns:
            None
        """
        
        
        self.data = data
        self.left = None
        self.right = None


class BinaryTree:
    
    
    """
    Description:
        A class to represent a binary tree.
        
    Parameter: 
        None

    Attributes:
        root (Node): The root of the binary tree.
    """

    def __init__(self):
        
        
        """
        Description:
            Initializes an empty binary tree.
            
        Parameter:
            None

        Returns:
            None
        """
        
        self.root = None



    def insert(self, data):
        
        
        """
        Description:
            Inserts a new node with the given data into the binary tree.

        Parameters:
            data (any): The data to be inserted into the binary tree.

        Returns:
            None
        """
        
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert_recursive(self.root, data)



    def _insert_recursive(self, node, data):
        
        
        """
        Description:
            Helper function to recursively insert a new node into the binary tree.

        Parameters:
            node (Node): The current node.
            data (any): The data to be inserted.

        Returns:
            None
        """
        
        
        if data < node.data:
            if node.left is None:
                node.left = Node(data)
            else:
                self._insert_recursive(node.left, data)
        else:
            if node.right is None:
                node.right = Node(data)
            else:
                self._insert_recursive(node.right, data)



    def inorder_traversal(self, node):
        
        """
        Description:
            Performs an inorder traversal of the binary tree.

        Parameters:
            node (Node): The current node.

        Returns:
            None
        """
        
        if node:
            self.inorder_traversal(node.left)
            print(node.data, end=" ")
            self.inorder_traversal(node.right)
            
            
    def search(self, data):
        
        
        """
        Description:
            Searches for a node with the given data in the binary tree.

        Parameters:
            data (any): The data to search for.

        Returns:
            bool: True if the data is found, False otherwise.
        """
        
        
        return self._search_recursive(self.root, data)



    def _search_recursive(self, node, data):
        
        
        """
        Description:
            Helper function to recursively search for a node in the binary tree.

        Parameters:
            node (Node): The current node.
            data (any): The data to search for.

        Returns:
            bool: True if the data is found, False otherwise.
        """
        
        if node is None:
            return False
        if data == node.data:
            return True
        elif data < node.data:
            return self._search_recursive(node.left, data)
        else:
            return self._search_recursive(node.right, data)      
        

 

def main():

    tree = BinaryTree()
    tree.insert(5)
    tree.insert(3)
    tree.insert(7)
    tree.insert(2)
    tree.insert(4)
    tree.insert(6)
    tree.insert(8)

    print("Inorder Traversal:")
    tree.inorder_traversal(tree.root)  

    
if __name__ == "__main__":
    main()
```

2.Finding Maximum Depth of Binary Tree

```python

def max_depth(node):
    
    """
    Description:
        Finds the maximum depth (height) of a binary tree.

    Parameters:
        node (Node): The root node of the binary tree.

    Returns:
        int: The maximum depth of the binary tree.
    """
    
    if node is None:
        return 0
    else:
        left_depth = max_depth(node.left)
        right_depth = max_depth(node.right)
        return max(left_depth, right_depth) + 1


def main():

    tree = BinaryTree()
    tree.insert(5)
    tree.insert(3)
    tree.insert(7)
    tree.insert(2)
    tree.insert(4)
    tree.insert(6)
    tree.insert(8)

    print("Maximum Depth of the Tree:", max_depth(tree.root))  # Output: 3


if __name__ == "__main__":
    main()
```
